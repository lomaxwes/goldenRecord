```python
import numpy as np
import pandas as pd
import shutil

df = pd.read_csv('ds_dirty_fin_202410041147.csv', delimiter=',',dtype = str)
df
```

```python
dfcl = df.dropna(subset = ['source_cd']).copy()
dfcl
```

```python
dfcl['source_cd'] = dfcl['source_cd'].str.upper()
df_sources = dfcl.groupby(['source_cd']).size().reset_index(name = 'records')
df_sources
```

```python
df_sources['columns'] = 0
```

```python
sources = list(dfcl.source_cd.unique())
```

```python
for rec in sources:
  df_sources.loc[df_sources['source_cd'] == rec,'columns'] = len(dfcl[dfcl.source_cd == rec].dropna(axis='columns',how='all').columns)

df_columns = list(dfcl.columns)
```

```python
df_columns.remove('client_id')
df_columns.remove('create_date')
df_columns.remove('update_date')
```

```python
def trim(str1):
  return str1.replace(' ', '')
```

```python
dfcl = dfcl.fillna('')
for rec in df_columns:
  dfcl[rec] = dfcl[rec].apply(trim)
  dfcl[rec] = dfcl[rec].str.upper()
```

```python
dfcl.drop_duplicates(subset=df_columns, inplace=True)
dfcl
```

```python
conditions = list()
choices = list(range(len(sources)))
```

```python
for rec in choices:
  conditions.append(dfcl['source_cd'] == sources[rec])
```

```python
dfcl['source_id'] = np.select(conditions, choices)
```

```python
df_sorted = dfcl.sort_values(by = ['source_id','client_inn'])
df_sorted
```
